```python
import pandas as pd
filename = '/content/drive/MyDrive/Coding Dojo PH/DS_Bootcamp/01 Data Science Fundamentals/Wk3 - Exploratory Data Visualization/Assignments/dataset/sales_predictions_2023.csv'

df_sales = pd.read_csv(filename)
df_sales.head()
```

# Histogram

```python
# Histograms shows the distribution of various features

import matplotlib.pyplot as plt

# Histogram of Item Weight
df_sales['Item_Weight'].hist(bins=50)
plt.title('Histogram of Item Weight')
plt.xlabel('Weight')
plt.ylabel('Number of Items')
plt.show()
```

```python
# Histogram of Item MRP
df_sales['Item_MRP'].hist(bins=50)
plt.title('Histogram of Item MRP (Maximum Retail Price)')
plt.xlabel('MRP')
plt.ylabel('Number of Items')
plt.show()
```

# Boxplot

```python
# Boxplots provides a graphical view of the central tendency and variability of the data, along with the presence of outliers.

import seaborn as sns

# Boxplot for Item_Outlet_Sales to see the distribution and outliers
sns.boxplot(x='Item_Outlet_Sales', data=df_sales)
plt.title('Boxplot of Item Outlet Sales')
plt.xlabel('Item Outlet Sales')
plt.show()
```

```python
# Boxplot for the numeric variables grouped by a categorical variable
sns.boxplot(x='Outlet_Size', y='Item_Outlet_Sales', data=df_sales)
plt.title('Boxplot of Item Outlet Sales by Outlet Size')
plt.xlabel('Outlet Size')
plt.ylabel('Item Outlet Sales')
plt.show()
```

# Heatmap

```python
# The Heatmap reveals the correlation coefficients between the numeric features

import numpy as np

# Calculate correlations
corr = df_sales.corr(numeric_only=True)

# Set up the matplotlib figure
plt.subplots(figsize=(11, 9))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, annot=True, fmt=".2f")
plt.title('Heatmap of Feature Correlations')
plt.show()
```
